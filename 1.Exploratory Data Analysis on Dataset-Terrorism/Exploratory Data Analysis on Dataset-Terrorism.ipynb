### Data Cleaning

```python
data=data[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
data.head()
```

```python
data.shape   #out total 135 columns we required 19 above columns to perform eda
```

### Renaming the columns

```python
data.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"Province/State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
data.head()
```

```python
data.dtypes
```

### Dealing with Null Values

```python
data.isna().sum()  #cheacking null values
```

```python
data['Killed']=data['Killed'].fillna(0)   #replacing null values with 0
data['Wounded']=data['Wounded'].fillna(0)
data['Casualties']=data['Killed']+data['Wounded']  #adding casualties column which showing no. of peoples which are killed and wounded
```

```python
data.head()
```

### Analysis

```python
data.describe()
```

### visualize year wise attacks

```python
attacks=data['Year'].value_counts().sort_index().to_frame().reset_index().rename(columns={'index':'Year','Year':'Attacks'}).set_index('Year')
attacks.head()
```

```python
attacks.plot(kind='bar',figsize=(15,6),fontsize=10,color='red',grid=True)
plt.title("Timeline of Attacks chart",fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

```python
attacks.describe(include='all')
```

```python
#Casualties in each Year
casualty_y=data[["Year","Casualties"]].groupby("Year").sum()
casualty_y.head()
```

```python
casualty_y.plot(kind='bar',figsize=(15,6),fontsize=10)
plt.xlabel("Year",fontsize=15)
plt.ylabel("No.of People Casualties",fontsize=15)
plt.title("Casualties in Each Year",fontsize=15)
plt.show()
```

###  Region wise Attacks

```python
reg=pd.crosstab(data['Year'],data['Region'])  #region
reg.head()
```

```python
reg.plot(kind='area',figsize=(15,8),fontsize=10)
plt.title("Region wise attacks",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

```python
#Total Terrorist Attacks in each Region from 1970-2017
reg_t=reg.transpose()   # getting transpose 
reg_t["Total"]=reg_t.sum(axis=1)
#regt['Total']
r_a=reg_t["Total"].sort_values(ascending=False)
r_a
```

```python
r_a.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
#Total Casualties (Killed + Wounded) in each Region
r_c=data[["Region","Casualties"]].groupby("Region").sum().sort_values(by="Casualties",ascending=False)
r_c
```

```python
r_c.plot(kind="bar",figsize=(15,6))
plt.title("Region wise Casualties",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

```python
r_k=data[["Region","Killed"]].groupby("Region").sum().sort_values(by="Killed",ascending=False)
r_k
```

```python
r_w=data[["Region","Wounded"]].groupby("Region").sum().sort_values(by="Wounded",ascending=False)
r_w
```

```python
#killed
r_k.plot(kind="bar",figsize=(15,6))
plt.title("People Killed in each Region",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.ylabel("Number of People Killed",fontsize=13)
#wounded
r_w.plot(kind="bar",figsize=(15,6))
plt.title("People Wounded in each Region")
plt.xlabel("Regions")
plt.ylabel("Number of People Wounded")
plt.show()
```
