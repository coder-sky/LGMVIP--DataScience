```python
#importing libraries
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

#### Loading image

```python
plt.figure(figsize=(10,7))
img = cv2.imread(r"S:\Projects\LGMVIP- DataScience\4.Image to Pencil Sketch with Python\avatar.jpg")
plt.imshow(img)  #here we got image with color get messed up from OpenCv to matplotlib (GBR image)
plt.show()
```

#### converting render image from OpenCv in matplotlib(GBR-->RGB)

```python
plt.figure(figsize=(10,7))
org_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #original image
plt.imshow(org_img)
plt.show()
```

#### converting original image to grey scale image

```python
plt.figure(figsize=(10,7))
gray_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img,cmap='gray')
plt.show()
```

#### invert the gray scale image

```python
plt.figure(figsize=(10,7))
inverted_img = 255 - gray_img
plt.imshow(inverted_img,cmap='gray')
plt.show()
```

#### blur the image by using the Gaussian Function

```python
blurred = cv2.GaussianBlur(inverted_img, (21, 21), 0)
```

#### pencil sketch

```python
plt.figure(figsize=(10,10))
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_img, inverted_blurred, scale=256.0)
plt.imshow(pencil_sketch,cmap='gray')
plt.show()
```

### original and pencil image comparision

```python
plt.figure(figsize=(25,10))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(org_img)
plt.subplot(1,2,2)
plt.title("Pencil Sketch")
plt.imshow(pencil_sketch,cmap='gray')
plt.show()
```
